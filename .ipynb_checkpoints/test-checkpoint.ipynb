```python
# ============================================================================
# 3D Model Retrieval Analysis with Rotation Invariance Testing
# Jupyter Notebook Version - Individual Functions
# ============================================================================

# Cell 1: Import Libraries and Setup
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import random
from mpl_toolkits.mplot3d import Axes3D
import trimesh
from scipy.spatial.transform import Rotation as R
from retrieval_system import HashRetrievalSystem

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

# Configure matplotlib for better plots
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['font.size'] = 10
```

```python
# ============================================================================
# Cell 2: Helper Functions
# ============================================================================

def apply_random_rotation(points):
    """Apply random 3D rotation to point cloud"""
    # Generate random rotation angles for x, y, z axes
    angles = np.random.uniform(0, 2*np.pi, 3)
    rotation = R.from_euler('xyz', angles)
    rotated_points = rotation.apply(points)
    return rotated_points

def load_and_rotate_model(model_path, n_points=1024, apply_rotation=True):
    """Load a 3D model, convert to point cloud, and apply random rotation"""
    try:
        mesh = trimesh.load_mesh(model_path)
        if hasattr(mesh, 'sample'):
            points = mesh.sample(n_points)
        else:
            points = mesh.vertices
            if len(points) > n_points:
                indices = np.random.choice(len(points), n_points, replace=False)
                points = points[indices]
        
        # Apply random rotation if requested
        if apply_rotation:
            points = apply_random_rotation(points)
            
        return points
    except Exception as e:
        print(f"Error loading {model_path}: {e}")
        return None

def get_category_models(database_path, target_category, n_models=5):
    """Get random models from a specific category"""
    category_path = os.path.join(database_path, target_category)
    if not os.path.exists(category_path):
        return []
    
    model_files = [f for f in os.listdir(category_path) 
                  if f.lower().endswith(('.off', '.obj', '.ply'))]
    
    # Randomly select n_models
    selected_files = random.sample(model_files, min(n_models, len(model_files)))
    return [os.path.join(category_path, f) for f in selected_files]

def get_dissimilar_models(database_path, exclude_category, n_models=5):
    """Get random models from categories other than the excluded one"""
    all_categories = [d for d in os.listdir(database_path) 
                     if os.path.isdir(os.path.join(database_path, d)) and d != exclude_category]
    
    dissimilar_models = []
    models_per_category = max(1, n_models // len(all_categories))
    
    for category in all_categories:
        category_models = get_category_models(database_path, category, models_per_category)
        dissimilar_models.extend(category_models)
        if len(dissimilar_models) >= n_models:
            break
    
    return random.sample(dissimilar_models, min(n_models, len(dissimilar_models)))

def compute_hash_with_rotation(retrieval_system, model_path):
    """Compute hash code for a model with random rotation applied"""
    rotated_points = load_and_rotate_model(model_path, apply_rotation=True)
    if rotated_points is None:
        return None
    
    # Compute hash using the retrieval system
    return retrieval_system.compute_hash(rotated_points)

print("Helper functions defined successfully!")
```

```python
# ============================================================================
# Cell 3: Data Collection Functions
# ============================================================================

def collect_evaluation_data(retrieval_system, query_path, database_path, n_similar=5, n_dissimilar=5):
    """
    Collect evaluation data for query model with similar and dissimilar models
    """
    # Get query information
    query_category = os.path.basename(os.path.dirname(query_path))
    query_name = os.path.splitext(os.path.basename(query_path))[0]
    
    print(f"Processing query: {query_category}/{query_name}")
    
    # Compute query hash with rotation
    query_hash = compute_hash_with_rotation(retrieval_system, query_path)
    if query_hash is None:
        print("Failed to process query model")
        return None
    
    # Get similar and dissimilar models
    print(f"Finding {n_similar} similar models from category: {query_category}")
    similar_models = get_category_models(database_path, query_category, n_similar)
    
    print(f"Finding {n_dissimilar} dissimilar models from other categories")
    dissimilar_models = get_dissimilar_models(database_path, query_category, n_dissimilar)
    
    # Prepare evaluation data
    evaluation_data = []
    
    # Add query model
    evaluation_data.append({
        'Type': 'Query',
        'Category': query_category,
        'Model_Name': query_name,
        'Full_Path': query_path,
        'Hash_Code': query_hash,
        'Similarity': 1.0,  # Perfect similarity with itself
        'Is_Similar': True
    })
    
    # Process similar models
    print("Processing similar models...")
    for i, model_path in enumerate(similar_models):
        model_hash = compute_hash_with_rotation(retrieval_system, model_path)
        if model_hash is not None:
            similarity = retrieval_system.compute_similarity(query_hash, model_hash)
            category = os.path.basename(os.path.dirname(model_path))
            model_name = os.path.splitext(os.path.basename(model_path))[0]
            
            evaluation_data.append({
                'Type': f'Similar_{i+1}',
                'Category': category,
                'Model_Name': model_name,
                'Full_Path': model_path,
                'Hash_Code': model_hash,
                'Similarity': similarity,
                'Is_Similar': True
            })
            print(f"  Similar {i+1}: {category}/{model_name} - Similarity: {similarity:.4f}")
    
    # Process dissimilar models
    print("Processing dissimilar models...")
    for i, model_path in enumerate(dissimilar_models):
        model_hash = compute_hash_with_rotation(retrieval_system, model_path)
        if model_hash is not None:
            similarity = retrieval_system.compute_similarity(query_hash, model_hash)
            category = os.path.basename(os.path.dirname(model_path))
            model_name = os.path.splitext(os.path.basename(model_path))[0]
            
            evaluation_data.append({
                'Type': f'Dissimilar_{i+1}',
                'Category': category,
                'Model_Name': model_name,
                'Full_Path': model_path,
                'Hash_Code': model_hash,
                'Similarity': similarity,
                'Is_Similar': False
            })
            print(f"  Dissimilar {i+1}: {category}/{model_name} - Similarity: {similarity:.4f}")
    
    return evaluation_data

print("Data collection functions defined successfully!")
```

```python
# ============================================================================
# Cell 4: Visualization Functions
# ============================================================================

def plot_model_3d(ax, model_path, title, color, size=2):
    """Plot a 3D model with rotation applied"""
    points = load_and_rotate_model(model_path, n_points=5000, apply_rotation=True)
    if points is None:
        ax.text(0.5, 0.5, 0.5, 'Failed to load', ha='center', va='center')
        return
    
    ax.scatter(points[:, 0], points[:, 1], points[:, 2], 
              c=color, alpha=0.7, s=size)
    ax.set_title(title, fontsize=8, fontweight='bold')
    
    # Clean up axes
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    
    # Set equal aspect ratio
    max_range = np.array([points[:,0].max()-points[:,0].min(),
                         points[:,1].max()-points[:,1].min(),
                         points[:,2].max()-points[:,2].min()]).max() / 2.0
    mid_x = (points[:,0].max()+points[:,0].min()) * 0.5
    mid_y = (points[:,1].max()+points[:,1].min()) * 0.5
    mid_z = (points[:,2].max()+points[:,2].min()) * 0.5
    
    ax.set_xlim(mid_x - max_range, mid_x + max_range)
    ax.set_ylim(mid_y - max_range, mid_y + max_range)
    ax.set_zlim(mid_z - max_range, mid_z + max_range)

def create_info_text(evaluation_data):
    """Create detailed information text for visualization"""
    query_data = [d for d in evaluation_data if d['Type'] == 'Query'][0]
    similar_data = [d for d in evaluation_data if d['Is_Similar'] and d['Type'] != 'Query']
    dissimilar_data = [d for d in evaluation_data if not d['Is_Similar']]
    
    # Calculate statistics
    similar_similarities = [d['Similarity'] for d in similar_data]
    dissimilar_similarities = [d['Similarity'] for d in dissimilar_data]
    
    avg_similar = np.mean(similar_similarities) if similar_similarities else 0
    avg_dissimilar = np.mean(dissimilar_similarities) if dissimilar_similarities else 0
    
    info_lines = [
        "ROTATION-INVARIANT RETRIEVAL EVALUATION RESULTS",
        "" * 70,
        f"Query Model: {query_data['Category']}/{query_data['Model_Name']}",
        f"Query Hash: {''.join(map(str, query_data['Hash_Code']))[:32]}...",
        "",
        f"SIMILAR MODELS (Same Category: {query_data['Category']}):",
        f"  Count: {len(similar_data)} | Avg Similarity: {avg_similar:.4f}",
        "",
        f"DISSIMILAR MODELS (Different Categories):",
        f"  Count: {len(dissimilar_data)} | Avg Similarity: {avg_dissimilar:.4f}",
        "",
        f"DISCRIMINATION SCORE: {avg_similar - avg_dissimilar:.4f}",
        f"(Higher values indicate better discrimination between similar/dissimilar models)",
        "",
        "Note: All models were randomly rotated before hash computation to test rotation invariance."
    ]
    
    return "\n".join(info_lines)

def format_hash_code(hash_code, length=12):
    """Format hash code for display (show first few and last few bits)"""
    hash_str = ''.join(map(str, hash_code))
    if len(hash_str) > length * 2:
        return hash_str[:length] + '...' + hash_str[-length:]
    return hash_str
    
def create_comprehensive_visualization(evaluation_data, save_path=None, figsize=(18, 12)):
    """Create comprehensive visualization of similar and dissimilar models with hash codes"""
    
    # Separate data
    query_data = [d for d in evaluation_data if d['Type'] == 'Query'][0]
    similar_data = [d for d in evaluation_data if d['Is_Similar'] and d['Type'] != 'Query']
    dissimilar_data = [d for d in evaluation_data if not d['Is_Similar']]
    
    # Create figure with better spacing
    fig = plt.figure(figsize=figsize)
    
    # Format query hash code
    query_hash = format_hash_code(query_data['Hash_Code'])
    
    # Plot query model (larger, centered) - centered in first row
    ax_query = fig.add_subplot(4, 6, (4, 4), projection='3d')  # Centered position
    query_title = f"QUERY\n{query_data['Category']}/{query_data['Model_Name']}\nHash: {query_hash}"
    plot_model_3d(ax_query, query_data['Full_Path'], query_title, 'blue', size=0.2)
    
    # Plot similar models - second row, centered (3 models)
    for i, data in enumerate(similar_data[:3]):
        ax = fig.add_subplot(4, 6, 8 + i + 1, projection='3d')  # positions 9, 10, 11 (centered)
        hash_code = format_hash_code(data['Hash_Code'])
        title = f"SIMILAR #{i+1}\n{data['Category']}/{data['Model_Name']}\nSim: {data['Similarity']:.3f}\nHash: {hash_code}"
        plot_model_3d(ax, data['Full_Path'], title, 'green', size=0.2)
    
    # Plot dissimilar models - third row, centered (3 models)
    for i, data in enumerate(dissimilar_data[:3]):
        ax = fig.add_subplot(4, 6, 14 + i + 1, projection='3d')  # positions 15, 16, 17 (centered)
        hash_code = format_hash_code(data['Hash_Code'])
        title = f"DISSIMILAR #{i+1}\n{data['Category']}/{data['Model_Name']}\nSim: {data['Similarity']:.3f}\nHash: {hash_code}"
        plot_model_3d(ax, data['Full_Path'], title, 'orange', size=0.2)
    
    # Add legend - closer to the plots
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', 
               markersize=12, label='Query Model'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='green', 
               markersize=10, label='Similar Models (Same Category)'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', 
               markersize=10, label='Dissimilar Models (Different Categories)')
    ]
    
    # Position the legend higher up and centered
    fig.legend(handles=legend_elements, loc='lower center', 
               bbox_to_anchor=(0.5, 0.1), ncol=3, framealpha=1)
    
    # Adjust spacing - reduced bottom margin, adjusted hspace
    plt.tight_layout()
    plt.subplots_adjust(bottom=0,  # Reduced from 0.15
                       top=0.92,     # Reduced from 0.88
                       left=0.05, 
                       right=0.95, 
                       hspace=0.3,   # Reduced vertical spacing between rows
                       wspace=0.15)  # Slightly increased horizontal spacing
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Comprehensive visualization saved to: {save_path}")
    
    plt.show()
```

```python
# ============================================================================
# Cell 5: Table Creation Functions
# ============================================================================

def generate_multi_query_evaluation_data(retrieval_system, database_path, n_results=20, similar_ratio=0.5):
    """
    Generate evaluation data with multiple different query models
    
    Args:
        retrieval_system: The hash retrieval system
        database_path: Path to the database
        n_results: Total number of rows to generate
        similar_ratio: Ratio of similar models (0.0 to 1.0)
    """
    # Get all available categories
    all_categories = [d for d in os.listdir(database_path) 
                     if os.path.isdir(os.path.join(database_path, d))]
    
    if len(all_categories) < 2:
        print("Need at least 2 categories for evaluation")
        return None
    
    evaluation_data = []
    n_similar = int(n_results * similar_ratio)
    n_dissimilar = n_results - n_similar
    
    print(f"Generating {n_results} evaluation pairs ({n_similar} similar, {n_dissimilar} dissimilar)")
    
    # Generate similar pairs (same category)
    for i in range(n_similar):
        # Randomly select a category
        category = random.choice(all_categories)
        category_models = get_category_models(database_path, category, 10)  # Get more options
        
        if len(category_models) < 2:
            continue
            
        # Randomly select two different models from the same category
        query_model, target_model = random.sample(category_models, 2)
        
        # Compute hashes and similarity
        query_hash = compute_hash_with_rotation(retrieval_system, query_model)
        target_hash = compute_hash_with_rotation(retrieval_system, target_model)
        
        if query_hash is not None and target_hash is not None:
            similarity = retrieval_system.compute_similarity(query_hash, target_hash)
            
            query_category = os.path.basename(os.path.dirname(query_model))
            query_name = os.path.splitext(os.path.basename(query_model))[0]
            target_category = os.path.basename(os.path.dirname(target_model))
            target_name = os.path.splitext(os.path.basename(target_model))[0]
            
            evaluation_data.append({
                'Query_Model': f"{query_category}/{query_name}",
                'Database_Model': f"{target_category}/{target_name}",
                'Similarity_Score': similarity,
                'Category_Match': 'Yes',
                'Is_Similar': True
            })
            print(f"Similar pair {i+1}: {similarity:.4f}")
    
    # Generate dissimilar pairs (different categories)
    for i in range(n_dissimilar):
        # Randomly select two different categories
        if len(all_categories) < 2:
            continue
            
        query_category, target_category = random.sample(all_categories, 2)
        
        query_models = get_category_models(database_path, query_category, 5)
        target_models = get_category_models(database_path, target_category, 5)
        
        if not query_models or not target_models:
            continue
            
        # Randomly select one model from each category
        query_model = random.choice(query_models)
        target_model = random.choice(target_models)
        
        # Compute hashes and similarity
        query_hash = compute_hash_with_rotation(retrieval_system, query_model)
        target_hash = compute_hash_with_rotation(retrieval_system, target_model)
        
        if query_hash is not None and target_hash is not None:
            similarity = retrieval_system.compute_similarity(query_hash, target_hash)
            
            query_name = os.path.splitext(os.path.basename(query_model))[0]
            target_name = os.path.splitext(os.path.basename(target_model))[0]
            
            evaluation_data.append({
                'Query_Model': f"{query_category}/{query_name}",
                'Database_Model': f"{target_category}/{target_name}",
                'Similarity_Score': similarity,
                'Category_Match': 'No',
                'Is_Similar': False
            })
            print(f"Dissimilar pair {i+1}: {similarity:.4f}")
    
    return evaluation_data

def create_evaluation_table(evaluation_data, save_path=None):
    """Create and save evaluation results as a table"""
    
    # Prepare table data
    table_data = []
    query_data = [d for d in evaluation_data if d['Type'] == 'Query'][0]
    
    for data in evaluation_data:
        # Format hash code (show first 16 and last 16 bits)
        hash_str = ''.join(map(str, data['Hash_Code']))
        if len(hash_str) > 32:
            formatted_hash = hash_str[:16] + '...' + hash_str[-16:]
        else:
            formatted_hash = hash_str
        
        table_data.append({
            'Query_Model': f"{query_data['Category']}/{query_data['Model_Name']}",
            'Database_Model': f"{data['Category']}/{data['Model_Name']}",
            'Model_Type': data['Type'],
            'Hash_Code': formatted_hash,
            'Similarity_Score': f"{data['Similarity']:.4f}",
            'Category_Match': 'Yes' if data['Is_Similar'] else 'No'
        })
    
    # Create DataFrame
    df = pd.DataFrame(table_data)
    
    # Save as CSV if path provided
    if save_path:
        csv_path = save_path.replace('.png', '.csv').replace('.pdf', '.csv')
        df.to_csv(csv_path, index=False)
        print(f"Evaluation table saved to: {csv_path}")
    
    return df

def create_multi_query_table(multi_query_data, save_path=None):
    """Create table from multi-query evaluation data"""
    
    # Create DataFrame directly from the multi-query data
    df_data = []
    for data in multi_query_data:
        df_data.append({
            'Query_Model': data['Query_Model'],
            'Database_Model': data['Database_Model'],
            'Similarity_Score': f"{data['Similarity_Score']:.4f}",
            'Category_Match': data['Category_Match']
        })
    
    df = pd.DataFrame(df_data)
    
    # Save as CSV if path provided
    if save_path:
        csv_path = save_path.replace('.png', '.csv').replace('.pdf', '.csv')
        df.to_csv(csv_path, index=False)
        print(f"Multi-query evaluation table saved to: {csv_path}")
    
    return df

def create_visual_table(df, save_path=None):
    """Create visual table from DataFrame"""
    
    # Create visual table
    fig, ax = plt.subplots(figsize=(14, max(8, len(df) * 0.4)))
    ax.axis('tight')
    ax.axis('off')
    
    # Create table
    table = ax.table(cellText=df.values, colLabels=df.columns, 
                    loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1.2, 1.8)
    
    # Style the table
    header_color = '#4CAF50'
    similar_color = '#E8F5E8'
    dissimilar_color = '#FFE8E8'
    
    # Color code rows based on Category_Match
    for i in range(len(df)):
        row_data = df.iloc[i]
        if row_data['Category_Match'] == 'Yes':
            color = similar_color
        else:
            color = dissimilar_color
        
        for j in range(len(df.columns)):
            table[(i+1, j)].set_facecolor(color)
    
    # Style header
    for j in range(len(df.columns)):
        table[(0, j)].set_facecolor(header_color)
        table[(0, j)].set_text_props(weight='bold', color='white')
    
    plt.title('Multi-Query Rotation-Invariant Retrieval Evaluation Results', 
             fontsize=14, fontweight='bold', pad=20)
    
    # Add color legend
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor=similar_color, label='Similar Models (Same Category)'),
        Patch(facecolor=dissimilar_color, label='Dissimilar Models (Different Categories)')
    ]
    ax.legend(handles=legend_elements, loc='upper center', ncol=2)
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Visual table saved to: {save_path}")
    
    plt.show()
    return fig

print("Table creation functions defined successfully!")
```

```python
# ============================================================================
# Cell 6: Statistics and Analysis Functions
# ============================================================================

def analyze_results(evaluation_data):
    """Analyze and print detailed statistics from evaluation results"""
    
    query_data = [d for d in evaluation_data if d['Type'] == 'Query'][0]
    similar_data = [d for d in evaluation_data if d['Is_Similar'] and d['Type'] != 'Query']
    dissimilar_data = [d for d in evaluation_data if not d['Is_Similar']]
    
    # Calculate statistics
    similar_similarities = [d['Similarity'] for d in similar_data]
    dissimilar_similarities = [d['Similarity'] for d in dissimilar_data]
    
    print("="*60)
    print("DETAILED ANALYSIS RESULTS")
    print("="*60)
    print(f"Query Model: {query_data['Category']}/{query_data['Model_Name']}")
    print()
    
    if similar_similarities:
        print("SIMILAR MODELS STATISTICS:")
        print(f"  Count: {len(similar_similarities)}")
        print(f"  Average Similarity: {np.mean(similar_similarities):.4f}")
        print(f"  Max Similarity: {np.max(similar_similarities):.4f}")
        print(f"  Min Similarity: {np.min(similar_similarities):.4f}")
        # print(f"  Std Deviation: {np.std(similar_similarities):.4f}")
        print()
    
    if dissimilar_similarities:
        print("DISSIMILAR MODELS STATISTICS:")
        print(f"  Count: {len(dissimilar_similarities)}")
        print(f"  Average Similarity: {np.mean(dissimilar_similarities):.4f}")
        print(f"  Max Similarity: {np.max(dissimilar_similarities):.4f}")
        print(f"  Min Similarity: {np.min(dissimilar_similarities):.4f}")
        # print(f"  Std Deviation: {np.std(dissimilar_similarities):.4f}")
        print()
    
    if similar_similarities and dissimilar_similarities:
        discrimination_score = np.mean(similar_similarities) - np.mean(dissimilar_similarities)
        print("DISCRIMINATION ANALYSIS:")
        print(f"  Discrimination Score: {discrimination_score:.4f}")
        # print(f"  Interpretation: {'Excellent' if discrimination_score > 0.3 else 'Good' if discrimination_score > 0.1 else 'Needs Improvement'}")
        print()
    
    # Category breakdown for dissimilar models
    dissimilar_categories = [d['Category'] for d in dissimilar_data]
    unique_categories = list(set(dissimilar_categories))
    
    print("DISSIMILAR MODELS BY CATEGORY:")
    for category in unique_categories:
        category_data = [d for d in dissimilar_data if d['Category'] == category]
        category_sims = [d['Similarity'] for d in category_data]
        print(f"  {category}: {len(category_data)} models, Avg Sim: {np.mean(category_sims):.4f}")

def create_similarity_distribution_plot(evaluation_data, save_path=None):
    """Create histogram showing similarity score distributions"""
    
    similar_data = [d for d in evaluation_data if d['Is_Similar'] and d['Type'] != 'Query']
    dissimilar_data = [d for d in evaluation_data if not d['Is_Similar']]
    
    similar_similarities = [d['Similarity'] for d in similar_data]
    dissimilar_similarities = [d['Similarity'] for d in dissimilar_data]
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Similar models histogram
    ax1.hist(similar_similarities, bins=10, alpha=0.7, color='green', edgecolor='black')
    ax1.set_title('Similar Models\nSimilarity Score Distribution')
    ax1.set_xlabel('Similarity Score')
    ax1.set_ylabel('Frequency')
    ax1.axvline(np.mean(similar_similarities), color='red', linestyle='--', 
                label=f'Mean: {np.mean(similar_similarities):.3f}')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Dissimilar models histogram
    ax2.hist(dissimilar_similarities, bins=10, alpha=0.7, color='orange', edgecolor='black')
    ax2.set_title('Dissimilar Models\nSimilarity Score Distribution')
    ax2.set_xlabel('Similarity Score')
    ax2.set_ylabel('Frequency')
    ax2.axvline(np.mean(dissimilar_similarities), color='red', linestyle='--', 
                label=f'Mean: {np.mean(dissimilar_similarities):.3f}')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Similarity distribution plot saved to: {save_path}")
    
    plt.show()
    return fig

def analyze_multi_query_results(multi_query_data):
    """Analyze and print detailed statistics from multi-query evaluation results"""
    
    similar_data = [d for d in multi_query_data if d['Is_Similar']]
    dissimilar_data = [d for d in multi_query_data if not d['Is_Similar']]
    
    # Calculate statistics
    similar_similarities = [d['Similarity_Score'] for d in similar_data]
    dissimilar_similarities = [d['Similarity_Score'] for d in dissimilar_data]
    
    print("="*60)
    print("DETAILED MULTI-QUERY ANALYSIS RESULTS")
    print("="*60)
    print(f"Total Test Pairs: {len(multi_query_data)}")
    print()
    
    if similar_similarities:
        print("SIMILAR MODELS STATISTICS:")
        print(f"  Count: {len(similar_similarities)}")
        print(f"  Average Similarity: {np.mean(similar_similarities):.4f}")
        # print(f"  Max Similarity: {np.max(similar_similarities):.4f}")
        # print(f"  Min Similarity: {np.min(similar_similarities):.4f}")
        # print(f"  Std Deviation: {np.std(similar_similarities):.4f}")
        print()
    
    if dissimilar_similarities:
        print("DISSIMILAR MODELS STATISTICS:")
        print(f"  Count: {len(dissimilar_similarities)}")
        print(f"  Average Similarity: {np.mean(dissimilar_similarities):.4f}")
        # print(f"  Max Similarity: {np.max(dissimilar_similarities):.4f}")
        # print(f"  Min Similarity: {np.min(dissimilar_similarities):.4f}")
        # print(f"  Std Deviation: {np.std(dissimilar_similarities):.4f}")
        print()
    
    if similar_similarities and dissimilar_similarities:
        discrimination_score = np.mean(similar_similarities) - np.mean(dissimilar_similarities)
        print("DISCRIMINATION ANALYSIS:")
        print(f"  Discrimination Score: {discrimination_score:.4f}")
        # print(f"  Interpretation: {'Excellent' if discrimination_score > 0.3 else 'Good' if discrimination_score > 0.1 else 'Needs Improvement'}")
        print()
    
    # Category analysis for query models
    query_categories = [d['Query_Model'].split('/')[0] for d in multi_query_data]
    unique_query_categories = list(set(query_categories))
    
    print("QUERY MODELS BY CATEGORY:")
    for category in unique_query_categories:
        category_count = query_categories.count(category)
        print(f"  {category}: {category_count} queries")
    print()
```

```python
# ============================================================================
# Cell 7: Main Execution Function
# ============================================================================

def run_complete_analysis(retrieval_system, query_path, database_path, output_dir="thesis_outputs"):
    """
    Run complete analysis including visualization, table, and statistics
    """
    # Create output directory
    os.makedirs(output_dir, exist_ok=True)
    
    # Get query info for file naming
    query_category = os.path.basename(os.path.dirname(query_path))
    query_name = os.path.splitext(os.path.basename(query_path))[0]
    base_filename = f"{query_category}_{query_name}"
    
    print(f"Starting complete analysis for: {query_category}/{query_name}")
    print("="*60)
    
    # Step 1: Collect evaluation data
    print("Step 1: Collecting evaluation data...")
    evaluation_data = collect_evaluation_data(retrieval_system, query_path, database_path, 
                                            n_similar=5, n_dissimilar=5)
    
    if evaluation_data is None:
        print("Failed to collect evaluation data!")
        return None
    
    # Step 2: Create comprehensive visualization
    print("\nStep 2: Creating comprehensive visualization...")
    viz_path = os.path.join(output_dir, f"{base_filename}_retrieval_analysis.png")
    fig_viz = create_comprehensive_visualization(evaluation_data, save_path=viz_path)
    
    # Step 3: Create evaluation table
    print("\nStep 3: Creating evaluation table...")
    df = create_evaluation_table(evaluation_data)
    
    # Save CSV
    csv_path = os.path.join(output_dir, f"{base_filename}_evaluation_table.csv")
    df.to_csv(csv_path, index=False)
    print(f"CSV table saved to: {csv_path}")
    
    # Create visual table
    table_path = os.path.join(output_dir, f"{base_filename}_evaluation_table.png")
    fig_table = create_visual_table(df, save_path=table_path)
    
    # Step 4: Create similarity distribution plot
    print("\nStep 4: Creating similarity distribution analysis...")
    dist_path = os.path.join(output_dir, f"{base_filename}_similarity_distribution.png")
    fig_dist = create_similarity_distribution_plot(evaluation_data, save_path=dist_path)
    
    # Step 5: Analyze results
    print("\nStep 5: Analyzing results...")
    analyze_results(evaluation_data)
    
    print("\n" + "="*60)
    print("ANALYSIS COMPLETE!")
    print(f"All outputs saved to: {output_dir}")
    print("="*60)
    
    return {
        'evaluation_data': evaluation_data,
        'dataframe': df,
        'figures': {
            'visualization': fig_viz,
            'table': fig_table,
            'distribution': fig_dist
        }
    }
```
